```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("all_data.csv")
```

```python
#rename Life Expectancy column
#rename United States of America to United States for later merge with worldpop data
df.rename(columns={"Life expectancy at birth (years)":"lifex"}, inplace=True)
df["Country"] = df["Country"].replace("United States of America", "United States")
```

```python
#survey data
print(df.head())
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
#Scope of the project:
#1.) Display descriptive statistics; 
# A.) Which countries were surveyed?
# B.) What are the basic descriptive statistics of life expectancies at birth and GDP in the whole data set?
# C.) Visualize, by country, both statistics.

#2.) Display correlations;
# A.) Controlling for country, regress life expectancy on GDP
# B.) Controlling for GDP, regress life expectancy on calendar year

#3.) Identify outliers for further analysis and suggest avenues of investigation
# - Are there countries that deviate significantly from trends in life expectancy based on GDP or Calendar Year? Why might this be?
```

```python
#Descriptive Statistics 
print(df["Country"].unique())
print(df[["GDP", "lifex"]].describe())
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Country", y="GDP")
plt.xticks(rotation=45)
plt.title("GDP by Country")
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Country", y="lifex")
plt.xticks(rotation=45)
plt.title("life expectancy by Country")
plt.show()
plt.clf()
```

```python
#Integrating world population data into dataframe to calculate per-capita GDP for better insight into life expectancy.
df2 = pd.read_csv("wbpopdata.csv", skiprows=4)
print(df2.head())
print(df2.shape)
print(df2.dtypes)
print(df2.isnull().sum())
```

```python
# Only keep what we need
pop_df = df2[["Country Name"] + [str(year) for year in range(2000, 2024)]]

# Melt it from wide to long format
pop_long = pop_df.melt(id_vars="Country Name", 
                       var_name="Year", 
                       value_name="Population")

# Convert year to integer for merging
pop_long["Year"] = pop_long["Year"].astype(int)
```

```python
#attempt to match country names
df["Country"] = df["Country"].str.strip()
pop_long["Country Name"] = pop_long["Country Name"].str.strip()
```

```python
#do the merge
merged = df.merge(pop_long, how="left", 
                   left_on=["Country", "Year"], 
                   right_on=["Country Name", "Year"])
merged.drop(columns=["Country Name"], inplace=True)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=merged, x="Country", y="GDP per capita")
plt.xticks(rotation=45)
plt.title("Per capita GDP by Country")
plt.show()
plt.clf()
```

```python
#do regressions
import statsmodels.formula.api as smf
merged.columns = merged.columns.str.replace(" ","_")
model = smf.ols("lifex ~ GDP_per_capita + C(Year)", data=merged).fit()
print(model.summary())
```

```python
model = smf.ols("lifex ~ Year + C(Country)", data=merged).fit()
print(model.summary())
```

```python
#Conclusions
#Initial regression yielded confusing negative coefficient due to multicollinearity of per-capita GDP and Country; all countries had high GDP per capita
#for EVERY observation. Regression of life expectancy on GDP per capita, controlling only for year, yielded a statistically significant, but modest, 
#positive correlation. However, regressing life expectancy on year, controlling for country, yielded extremely significant and pronounced results. 
#across countries, across the 16 year period covered by the data, an additional year passing increased life expectancy by 4 months. 
#This could be due to technological advances, geopolitical stability, or other effects. 
```
